```python
# code

import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('fakeData.txt', delimiter=',', skip_header=1)

#print(data)

x = data[:, 0]
y = data[:, 1]
bad_flag = data[:, 2]


good_data = (bad_flag != -1)
bad_data = (bad_flag == -1)


x_good = x[good_data]
y_good = y[good_data]

x_bad = x[bad_data]
y_bad = y[bad_data]


plt.scatter(x_good, y_good, color='blue', label='Good Data', s=15)
plt.scatter(x_bad, y_bad, color='red', marker='X', label='Bad Data', s=15)


coefficients = np.polyfit(x_good, y_good, 2)
polynomial = np.poly1d(coefficients)


x_line = np.linspace(min(x_good), max(x_good), 100)
y_line = polynomial(x_line)


plt.plot(x_line, y_line, color='green', label='Fitted Polynomial')


plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Good vs Bad Data Polynomial Fit')
plt.show()
```
