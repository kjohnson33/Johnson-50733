Now we'll conveniently forget that we "knew" acceleration due to gravity for the sake of this exercise. 

We will however remember that objects in motion under the influence of a constant acceleration, in a vacuum, move according to the equation $ x = x_0 + v\cdot t $+$ a\cdot t^2 \over 2$.

We should be able to recover the acceleration due to gravity by finding the three coefficients of a second degree polynomial fit to our fake data (use `noisy`, not `y`).

- a) Write a 2nd degree polynomial function then use $\chi^2$ minimization to find the three best fit coefficients (25 pts)

```python
# code 

### part a: best fit coefficients

import copy
import numpy as np
import matplotlib.pyplot as plt


def polynomial(t, parame):
    x0, v, a = parame
    return x0 + (v * t) + (0.5 * a * t**2)


def chi_square(parame, t, y_data):
    y_model = polynomial(t, parame)
    return np.sum((y_data - y_model) ** 2)


def nelder_mead(f, x_start, t, y_data, step=0.1, max_iter=1000, verbose=False):
    
    dim = len(x_start)
    prev_best = f(x_start, t, y_data)
    no_improv = 0
    res = [[x_start, prev_best]]
    
    for i in range(dim):
        x = copy.copy(x_start)
        x[i] += step
        score = f(x, t, y_data)
        res.append([x, score])

    iters = 0
    while True:
        res.sort(key=lambda x: x[1])
        best = res[0][1]
        if verbose:
            print(f"Iteration {iters}, best score: {best}")
        
        iters += 1
        if iters >= max_iter:
            return res[0][0]

        # Check for no improvement
        if best < prev_best:
            no_improv = 0
            prev_best = best
        else:
            no_improv += 1
        if no_improv >= 10:
            return res[0][0]
        
        # Reflection
        x0 = np.mean([r[0] for r in res[:-1]], axis=0)
        xr = x0 + step * (x0 - res[-1][0])
        rscore = f(xr, t, y_data)
        
        if res[0][1] <= rscore < res[-2][1]:
            del res[-1]
            res.append([xr, rscore])
            continue
        
        # Expansion
        if rscore < res[0][1]:
            xe = x0 + 2 * step * (x0 - res[-1][0])
            escore = f(xe, t, y_data)
            if escore < rscore:
                del res[-1]
                res.append([xe, escore])
                continue
            else:
                del res[-1]
                res.append([xr, rscore])
                continue

        # Contraction
        xc = x0 + 0.5 * step * (x0 - res[-1][0])
        cscore = f(xc, t, y_data)
        if cscore < res[-1][1]:
            del res[-1]
            res.append([xc, cscore])
            continue
        
        # Reduction
        x1 = res[0][0]
        nres = []
        for tup in res:
            redx = x1 + 0.5 * step * (tup[0] - x1)
            score = f(redx, t, y_data)
            nres.append([redx, score])
        res = nres
    
    return res[0][0]


t = np.arange(0, 10, 0.1)  
y_data = 10 + (50 * t) - ((9.8 * t**2) / (2))  
noisy = y_data + np.random.randn(len(y_data)) * 10  


initial_guess = [10, 50, -9.8]  # x0, v, a
best_fit_parame = nelder_mead(chi_square, initial_guess, t, noisy)

x0, v, a = best_fit_parame


print("Best-fit parameters:")
print(f"x0 : {x0}")
print(f"v : {v}")
print(f"a : {a}")


### part b: Best Fit Poly

plt.scatter(t, noisy, label="Noisy Data", color='red')
plt.plot(t, polynomial(t, best_fit_parame), label="Best Fit", color='blue')
plt.xlabel('Time (sec)')
plt.ylabel('Height (m)')
plt.title('Best Fit Polynomial')
plt.legend()

plt.show()


### part c: acceleration due to gravity

gravity = best_fit_parame[2]
print(f"Acceleration due to gravity: {gravity:.2f} m/s²")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def line(x, m, b):
    return m*np.sin(x) + b

x = np.arange(0, 10, 0.1)
y_line = line(x, 2, 1)

# code


### part a: gradient descent


def chi_sq(y_obs, y_pred):
    return np.sum((y_obs - y_pred)**2)


def gradient_m(x, y_obs, m, b):
    
    return -2 * np.sum((y_obs - line(x, m, b)) * np.sin(x))

def gradient_b(x, y_obs, m, b):
    
    return -2 * np.sum(y_obs - line(x, m, b))


def gradient_descent(x, y_obs, m_init, b_init, step_size=0.001, itera=1000, tolerance=1e-6):
    m = m_init
    b = b_init
    chi_square_vals = []
    
    for iterations in range(itera):
        
        grad_m = gradient_m(x, y_obs, m, b)
        grad_b = gradient_b(x, y_obs, m, b)
        
        
        m -= step_size * grad_m
        b -= step_size * grad_b
        
        # chi-square calcs and store 
        y_pred = line(x, m, b)
        chi_square_val = chi_sq(y_obs, y_pred)
        chi_square_vals.append(chi_square_val)
        
        # progress and check if the chi-square becomes too large
#         if itera % 10 == 0:
#             print(f"iterations {itera}: m={m}, b={b}, chi-square={chi_square_val}")
        
        # If chi-squar extremely large stop
        if chi_square_val > 1e10:  
            print("Chi-square value is too large, now stopping.")
            break
    
    return m, b, chi_square_vals

# path track
def gradient_descent_path(x, y_obs, m_init, b_init, step_size=0.001, itera=1000):
    m = m_init
    b = b_init
    path = []
    
    for iterations in range(itera):
        
        grad_m = gradient_m(x, y_obs, m, b)
        grad_b = gradient_b(x, y_obs, m, b)
        
        
        m -= step_size * grad_m
        b -= step_size * grad_b
        
        # Store current path point
        path.append((m, b))
    
    return m, b, path


m_init = 0.0
b_init = 0.0
step_size = 0.001
itera = 1000

m_opt, b_opt, chi_square_vals = gradient_descent(x, y_line, m_init, b_init, step_size, itera)

print(f"Coefficient of m: {m_opt}, Coefficient of b: {b_opt}")


### part b: best fit

x = np.arange(0, 10, 0.1)
y_true = 2 * np.sin(x) + 1  


# original data and the best fit line plots
plt.scatter(x, y_line, label='Data', color='blue')
plt.plot(x, line(x, m_opt, b_opt), label='Best Fit Line', color='red', lw=4)
plt.plot(x, y_true, color='green', label='True function', lw=2.5)
plt.title('Best Fit Line Using Gradient Descent')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()


### part c: chi-square surface and path

# Plot the chi-square surface
m_vals = np.linspace(-7, 7, 100)
b_vals = np.linspace(-7, 7, 100)
M, B = np.meshgrid(m_vals, b_vals)

# Calculate chi-square values for each (m, b) coordinate pair
Z = np.zeros_like(M)
for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        y_pred = line(x, M[i, j], B[i, j])
        Z[i, j] = chi_sq(y_line, y_pred)

# Plot chi-square surface
fig = plt.figure(figsize=(8, 12))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(M, B, Z, cmap='viridis')

ax.set_xlabel('m')
ax.set_ylabel('b')
ax.set_zlabel('Chi-Square')
ax.set_title('Chi-Square Surface')

plt.show()


# path from gradient descent
m_opt, b_opt, path = gradient_descent_path(x, y_line, m_init, b_init, step_size, itera)


# combine
path_m, path_b = zip(*path)


# chi square value calcs  for each (m,b) on path
path_chi_square_vals = [chi_sq(y_line, line(x, m, b)) for m, b in zip(path_m, path_b)]


# path on chi-square surface
fig = plt.figure(figsize=(8, 12))
ax = fig.add_subplot(111, projection='3d')
#ax.plot_surface(M, B, Z, cmap='cividis')
#ax.plot_surface(M, B, Z, alpha=0.3) # transparency 
ax.plot_surface(M, B, Z, cmap='viridis', alpha = 0.5)


# descent path on chi-square surface 
ax.plot(path_m, path_b, path_chi_square_vals, color='red', marker='*', markersize=10)
ax.set_xlabel('m')
ax.set_ylabel('b')
ax.set_zlabel('Chi-Square')
ax.set_title('Chi-Square Surface with Gradient Descent Path')

plt.show()
```

```python
# code

### part a: scipy optimize minimize

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize, curve_fit


def line(x, m, b):
    return m * np.sin(x) + b


def chi_sq2(para, x, y_obs):
    m, b = para
    y_pred = line(x, m, b)
    return np.sum((y_obs - y_pred)**2)

x = np.arange(0, 10, 0.1)
y_scipy = line(x, 2, 1)

initial_gus = [0.0, 0.0]


result = minimize(chi_sq2, initial_gus, args=(x, y_scipy))

# Output parameters
m_otp, b_otp = result.x

print(f"Minimize: Optimal m: {m_otp}, Optimal b: {b_otp}")

# Data and best fit (minimize)
plt.scatter(x, y_scipy, label='Data', color='blue')
plt.plot(x, line(x, m_otp, b_otp), label='Best Fit Line', color='red', lw=4)
plt.plot(x, line(x, 2, 1), color='green', label='True function', lw=2.5)
plt.title('Best Fit Line Using Scipy Optimize Minimize')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()


### part b: scipy optimize curve fit


def line(x, m, b):
    return m * np.sin(x) + b

x = np.arange(0, 10, 0.1)
y_scipy2 = line(x, 2, 1)  


potp, pcov = curve_fit(line, x, y_scipy2, p0=[0, 0])  

# Output parameters 2
m_otp2, b_otp2 = potp

print(f"Curve fit: Optimal m: {m_otp}, Optimal b: {b_otp}")

# Data and the best fit (curvefit)
plt.scatter(x, y_scipy2, label='Data', color='blue')
plt.plot(x, line(x, m_otp2, b_otp2), label='Best Fit Line', color='red', lw=4)
plt.plot(x, line(x, 2, 1), color='green', label='True function', lw=2.5)
plt.title('Best Fit Line Using Curve Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```
